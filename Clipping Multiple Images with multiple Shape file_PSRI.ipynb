```python
import os
import arcpy
from arcpy.sa import ExtractByMask

# Set the input and output folders
input_folder = r'C:\Users\Marium kona\OneDrive - The University of Texas at Austin\MA thesis work\Analysis\Other Vegetation Indices\PSRI 30 Years (1990-2022)'
output_folder_1 = r'H:\MA Thesis\PSRI 30 Years Clipped (1990-2022)\Big Area'
output_folder_2 = r'H:\MA Thesis\PSRI 30 Years Clipped (1990-2022)\Jurisdiction'
output_folder_3 = r'H:\MA Thesis\PSRI 30 Years Clipped (1990-2022)\Neighborhood and City Council'
output_folder_4 = r'H:\MA Thesis\PSRI 30 Years Clipped (1990-2022)\City Council'

# Set the shapefile paths
shapefile_1 = r'G:\USA GIS Files\Boundaries_ City of Austin Council Districts\Bigger Boundary\Austin_Bigger_Boundary.shp'
shapefile_2 = r'G:\USA GIS Files\Boundaries_ City of Austin Council Districts\All boundaries\Austin Jurisdiction.shp'
shapefile_3 = r'G:\USA GIS Files\Boundaries_ City of Austin Council Districts\All boundaries\Austin_Council_Districts.shp'
shapefile_4 = r'G:\USA GIS Files\Boundaries_ City of Austin Council Districts\All boundaries\Austin_Council_Districts.shp'

# Check out the ArcGIS Spatial Analyst extension
arcpy.CheckOutExtension('Spatial')

# Set the overwriteOutput parameter to True
arcpy.env.overwriteOutput = True

# Loop through the files and clip them using the first shapefile
for year in range(1992, 2023):
    if year < 2012:
        input_file = os.path.join(input_folder, f'Austin_LT05_027039_PSRI_{year}.tif')
    else:
        input_file = os.path.join(input_folder, f'Austin_LC08_027039_PSRI_{year}.tif')
    output_file = os.path.join(output_folder_1, f'PSRI_Big_Area_{year}.tif')
    out_raster = ExtractByMask(input_file, shapefile_1)
    out_raster.save(output_file)
    print(f'Successfully clipped {input_file} using {shapefile_1} and saved to {output_file}')

# Loop through the files and clip them using the second shapefile
for year in range(1992, 2023):
    if year < 2012:
        input_file = os.path.join(input_folder, f'Austin_LT05_027039_PSRI_{year}.tif')
    else:
        input_file = os.path.join(input_folder, f'Austin_LC08_027039_PSRI_{year}.tif')
    output_file = os.path.join(output_folder_2, f'PSRI_Jurisdiction_{year}.tif')
    out_raster = ExtractByMask(input_file, shapefile_2)
    out_raster.save(output_file)
    print(f'Successfully clipped {input_file} using {shapefile_2} and saved to {output_file}')

# Loop through the files and clip them using the third shapefile
for year in range(1992, 2023):
    if year < 2012:
        input_file = os.path.join(input_folder, f'Austin_LT05_027039_PSRI_{year}.tif')
    else:
        input_file = os.path.join(input_folder, f'Austin_LC08_027039_PSRI_{year}.tif')
    output_file = os.path.join(output_folder_3, f'PSRI_Neighborhood_and_City_Council_{year}.tif')
    out_raster = ExtractByMask(input_file, shapefile_3)
    out_raster.save(output_file)
    print(f'Successfully clipped {input_file} using {shapefile_3} and saved to {output_file}')

# Loop through the files and clip them using the fourth shapefile
for year in range(1992, 2023):
    if year < 2012:
        input_file = os.path.join(input_folder, f'Austin_LT05_027039_PSRI_{year}.tif')
    else:
        input_file = os.path.join(input_folder, f'Austin_LC08_027039_PSRI_{year}.tif')
    output_file = os.path.join(output_folder_4, f'PSRI_City_Council_{year}.tif')
    out_raster = ExtractByMask(input_file, shapefile_4)
    out_raster.save(output_file)
    print(f'Successfully clipped {input_file} using {shapefile_4} and saved to {output_file}')

# Check in the ArcGIS Spatial Analyst extension
arcpy.CheckInExtension('Spatial')
```
